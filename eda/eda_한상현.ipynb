```python
df.columns = ['Exited', 'Age', 'Gender', 'Dependents', 'Education', 'Marital', 'Income', 'Card_type', 
              'Tenure', 'Product_Cnt', 'Inactive_Months', 'Contacts_Cnt', 'Credit_Limit','Revolv_Bal', 'Avg_OTB', 'Amt_Chng_Q4_Q1',
              'Trans_Amt', 'Trans_Cnt', 'Cnt_Chng_Q4_Q1', 'Avg_Util_Ratio']
```

```python
df = df[['Exited', 'Age', 'Gender', 'Dependents', 'Education', 'Marital', 'Income', 'Card_type',
         'Tenure', 'Product_Cnt', 'Inactive_Months', 'Contacts_Cnt', 'Credit_Limit','Revolv_Bal', 'Avg_OTB',
         'Avg_Util_Ratio', 'Trans_Amt', 'Trans_Cnt', 'Amt_Chng_Q4_Q1', 'Cnt_Chng_Q4_Q1']]
df.head()
```

---

### 1. Age

```python
plt.figure(figsize=(15,6))
plt.subplot(1, 2, 1)                
plt.title('나이 분포')
sns.histplot(x='Age', data=df)

plt.subplot(1, 2, 2)                
sns.histplot(x='Age', data=df, hue='Exited', kde=True)
plt.title('이탈유무와 나이')
plt.tight_layout()
plt.show()
```

### 2. Gender

```python
plt.figure(figsize=(15,6))
plt.subplot(1, 2, 1)                
plt.title('성별')
sns.countplot(x='Gender', data=df)

plt.subplot(1, 2, 2)                
sns.countplot(x='Gender', data=df, hue='Exited')
plt.title('이탈유무와 성별')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,6))              
plt.title('성별에 따른 나이 분포')
sns.histplot(x='Age', data=df, hue='Gender', kde=True)
plt.show()
```

### 3. Dependents

```python
plt.figure(figsize=(15,6))
plt.subplot(1, 2, 1)                
plt.title('부양가족수')
sns.countplot(x='Dependents', data=df)

plt.subplot(1, 2, 2)                
sns.countplot(x='Dependents', data=df, hue='Exited')
plt.title('이탈유무와 부양가족수')
plt.tight_layout()
plt.show()
```

```python
# 이탈유무와 부양가족수 비율 확인
cross_tab_prop = pd.crosstab(index=df['Dependents'], columns=df['Exited'], normalize="index")
cross_tab_prop
```

### 4. Education

```python
plt.figure(figsize=(15,6))
plt.subplot(1, 2, 1)                
plt.title('교육수준')
sns.countplot(x='Education', data=df)

plt.subplot(1, 2, 2)                
sns.countplot(x='Education', data=df, hue='Exited')
plt.title('이탈유무와 교육수준')
plt.tight_layout()
plt.show()
```

```python
# 교육수준과 이탈유무 비율 확인
cross_tab_prop = pd.crosstab(index=df['Education'], columns=df['Exited'], normalize="index")
cross_tab_prop
```

```python
# 성별에 따른 교육수준 비율 확인
# df_01 = df.copy()
# df_01['Exited'] = df_01['Exited'].map({'Existing Customer': 0, 'Attrited Customer': 1})
cross_tab_prop = pd.crosstab(index=df['Education'], columns=df['Gender'], normalize="index")
cross_tab_prop
```

```python
plt.figure(figsize=(15,6))              
plt.title('나이에 따른 교육 수준 분포')
sns.histplot(x='Age', data=df, hue='Education', kde=True)
plt.show()
```

### 5. Marital

```python
plt.figure(figsize=(15,6))
plt.subplot(1, 2, 1)                
plt.title('결혼유무')
sns.countplot(x='Marital', data=df)

plt.subplot(1, 2, 2)                
sns.countplot(x='Marital', data=df, hue='Exited')
plt.title('이탈유무와 결혼유무')
plt.tight_layout()
plt.show()
```

```python
# 성별에 따른 결혼유무
sns.catplot(x='Gender',data=df, kind='count',col='Marital');
```

```python
plt.figure(figsize=(15,6))              
plt.title('나이에 따른 결혼 분포')
sns.histplot(x='Age', data=df, hue='Marital', kde=True)
plt.show()
```
